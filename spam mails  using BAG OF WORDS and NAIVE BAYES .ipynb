```python
import pandas as pd
import numpy as np
# 1-> spam
# 0-> no spam
```

```python
df=pd.read_csv("mail_data.csv")
df.head()
```

```python
df.Category.value_counts()
```

```python
#making a new column spam where if spam 1 otherwise 0
df['spam']=df['Category'].apply(lambda x: 1 if x=='spam' else 0)
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,Y_train,Y_test=train_test_split(df.Message,df.spam,test_size=0.2)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn.feature_extraction.text import CountVectorizer

v=CountVectorizer()

X_train_cv=v.fit_transform(X_train.values)
X_train_cv
```

```python
X_train_cv.toarray()
```

```python
# to see vocab
v.get_feature_names_out()[1000:1006]
```

```python
dir(v)
```

```python
v.vocabulary_
```

```python
X_train_cv.shape    #i have 4457 email and each email is of 7675 size
```

```python
X_train_np=X_train_cv.toarray()
X_train_np[0][1771] # checking if in sentence at pos 0 has word no 1771 in vocab
```

```python
#NOW ML MODEL
from sklearn.naive_bayes import MultinomialNB

model=MultinomialNB()
model.fit(X_train_cv,Y_train)
```

```python
X_test_cv=v.transform(X_test)
#converting x test to numerical x test
```

```python
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score


Y_pred=model.predict(X_test_cv)

print(classification_report(Y_test,Y_pred))
```

```python
score=accuracy_score(Y_test,Y_pred)
print("accuracy is=>",score)
```
